# Мини задание 4

В этот раз нужно будет немного поиздеваться над старой доброй VGG16.

Вам нужно загрузить предобученную сеть VGG16 и попробовать ее оптимизировать.

Примените к сетке следующие оптимизации
- Динамическая квантизация для полносвязных и сверточных слоев
- Неструктурированный прунинг для полносвязных и сверточных слоев

Обе оптимизации должны быть применены к одной модели! За каждую оптимизацию вы получаете **+5** баллов.

После загрузки модели и после каждой оптимизации, посчитайте следующие параметры
- Количество активных весов в сети
- Качество по Accuracy на датасете https://www.kaggle.com/datasets/deeptrial/miniimagenet 
- Время, за которое считается метрика на этом датасете
- Размер сети, если сохранить ее на жесткий диск

Проверять мы будем работоспособность и корректность всех применненых оптимизаций, поэтому структурируйте свой код так, чтобы можно было запустить все ячейки последовательно и получить результат!

PS. Неоговоренные параметры для прунинга и квантизации вы вольны выбирать сами.
